```python
import ee

# Trigger the authentication flow.
ee.Authenticate()

# Initialize the library.
ee.Initialize(project='ee-zafar')
```

```python
Map = geemap.Map()
Map
```

```python
country = ee.FeatureCollection("USDOS/LSIB_SIMPLE/2017")
Iceland = country.filter(ee.Filter.eq("country_na", "Iceland"))

Map.addLayer(Iceland,{},'Iceland')
Map.centerObject(Iceland, 8)
Map
```

```python
Dataset = ee.ImageCollection("MODIS/061/MOD10A1").filterDate('2001-01-01', '2023-12-31')
Snowcover =  Dataset.mean().select('NDSI_Snow_Cover').clip(Iceland)

SnowCoverViz = {
    min:0.0,
    max:100,
    'palette': ['blue', 'yellow', 'orange', 'red']
}
Map.addLayer(Snowcover, SnowCoverViz, 'Snow Cover')
Map
```

```python
img = ee.Image.pixelArea().divide(1e6)
img
```

```python
Iceland_Area = img.reduceRegion(**{
  'reducer': ee.Reducer.sum(),
  'geometry': Iceland,
  'scale': 500,
  'maxPixels': 1e13
})

Iceland_Area
```

```python
pixel_area = Snowcover.divide(100).divide(1000000).multiply(ee.Image.pixelArea())
pixel_area
```

```python
Ice_Area = pixel_area.reduceRegion(**{
  'reducer': ee.Reducer.sum(),
  'geometry': Iceland,
  'scale': 500,
  'maxPixels': 1e13
})

Ice_Area
```
